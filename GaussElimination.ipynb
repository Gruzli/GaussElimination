```python
import numpy as np

def input_dialog():
    '''
    Fungsi untuk membuat dialog dengan user. User akan memasukkan koefisien dari persamaan
    yang dicari. Fungsi akan mengembalikan matrik persamaan dan jumlah variabel yang tidak diketahui.
    '''
    m = int(input("Masukkan jumlah variabel: "))
    a = [[0.0] * (m+1) for i in range(m)]
    for i in range(m):
        print("Masukkan koefisien dan hasil persamaan ke %d" % (i+1))
        for j in range(m+1):
            if j == m:
                b = float(input("Hasil persamaan: "))
                a[i][j] = b
            else:
                b = float(input("Koefisien ke-%d: " % (j+1)))
                a[i][j] = b
    a = np.array(a, dtype=float)
    return (a, m)

def forward_elimination(b, n):
    """
    Fungsi untuk melakukan eliminasi Gauss untuk sistem persamaan linear Ax=b.
    """
    for i in range(n):
        if b[i][i] == 0:
            continue
        for j in range(i+1, n):
            ratio = b[j][i] / b[i][i]
            for k in range(n+1):
                b[j][k] = b[j][k] - ratio * b[i][k]
    return b

def back_substitution(c, n):
    '''
    Fungsi untuk melakukan back substitution. Fungsi akan mengembalikan sebuah array yang berisi hasil
    penyelesaian persamaan.
    '''
    y = np.zeros(n)
    y[n-1] = c[n-1][n] / c[n-1][n-1]
    for i in range(n-2, -1, -1):
        y[i] = c[i][n]
        for j in range(i+1, n):
            y[i] = y[i] - c[i][j] * y[j]
        y[i] /= c[i][i]
    return y

# Pemanggilan fungsi input_dialog() untuk meminta input koefisien persamaan
matrix, m = input_dialog()

# Pemanggilan fungsi forward_elimination() untuk melakukan eliminasi Gauss
matrix = forward_elimination(matrix, m)

# Pemanggilan fungsi back_substitution() untuk menyelesaikan sistem persamaan
result = back_substitution(matrix, m)

# Menampilkan hasil penyelesaian sistem persamaan
variables = ['x', 'y', 'z'] # sesuaikan dengan jumlah variabel yang diminta pada input_dialog()
print("Hasil dari persamaan:")
print(f"Batu bata: Rp {result[0]:,.3f} /kg".replace(',','.'))
print(f"Semen: Rp {result[1]:,.3f} /kg".replace(',','.'))
print(f"Besi: Rp {result[2]:,.3f} /kg".replace(',','.'))
```
